```python
# read the file
with open("C:/Sunbeam Git/Machine Learning/Machine_Learning/Assignment/Datasets/book.csv", "r") as file:
    lines = file.readlines()
```

```python
# create a list to hold all transactions
transactions = []

# read the lines one by one and convert them into transactions
for line in lines:
    # replace the \n with empty character
    line = line.replace('\n', '')

    # split the line with comma to get all the items separated
    transactions.append(line.split(','))
```

```python
transactions
```

```python
len(transactions)
```

```python
from apyori import apriori

# find the association rules using apriori
rules = list(apriori(transactions))
```

```python
required_support = 0.09
required_confidence = 0.09
```

```python
for rule in rules:
    support = rule.support
    confidence = rule.ordered_statistics[0].confidence
    if (support >= required_support) and (confidence >= required_confidence):
        print(f"rule = {rule.items}, support = {support}, confidence = {confidence}")
```

```python
from mlxtend.preprocessing import TransactionEncoder

# create an encoder
encoder = TransactionEncoder()

# encode the transactions 
encoded_transactions = encoder.fit_transform(transactions)

# convert the encoded transactions into a dataframe
df = pd.DataFrame(encoded_transactions, columns=encoder.columns_)
```

```python
from mlxtend.frequent_patterns import apriori

# find the rules
rules = apriori(df, min_support=0.09, use_colnames=True)
```

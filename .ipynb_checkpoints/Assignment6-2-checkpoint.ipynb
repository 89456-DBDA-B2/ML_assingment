- Perform K-Means,Hierarchical and DBSCAN Clustering for the crime_data.csv data and identify the number of clusters formed and Visualise the results.
- Data Description:
- Murder -- Muder rates in different places of United States
- Assualt- Assualt rate in different places of United States
- UrbanPop - urban population in different places of United States
- Rape - Rape rate in different places of United States

## prerequisite 

```python
import warnings
warnings.filterwarnings(action='ignore')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv('../crime_data.csv')
df.head(3)
```

```python
df.isna().sum()
```

```python
df.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
wss=[]
from sklearn.cluster import KMeans
for k in range(2,10):
    kmean = KMeans(n_clusters=k)
    kmean = kmean.fit(df)
    wss.append(kmean.inertia_)
```

```python
plt.scatter(range(2,10),wss)
plt.plot(range(2,10),wss,color='orange')
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5)

kmeans = kmeans.fit(df)
```

```python
kmeans.labels_
```

```python
kmeans.n_iter_
```

```python
colors = ['red','black','orange','blue','brown']


for i in range (5):
    plt.scatter(df['Murder'][kmeans.labels_==i],df['Assault'][kmeans.labels_==i],color=colors[i])
```

```python
colors = ['red','black','orange','blue','brown']


for i in range (5):
    plt.scatter(df['Murder'][kmeans.labels_==i],df['UrbanPop'][kmeans.labels_==i],color=colors[i])
```

```python
colors = ['red','black','orange','blue','brown']


for i in range (5):
    plt.scatter(df['Murder'][kmeans.labels_==i],df['Rape'][kmeans.labels_==i],color=colors[i])
```

# Hirarchical 

```python
df1=pd.read_csv('../crime_data.csv')
df1.head(3)
```

```python
df1.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
from sklearn.cluster import AgglomerativeClustering

agglo = AgglomerativeClustering(n_clusters=5)
agglo = agglo.fit(df1)
```

```python
colors = ['red','black','orange','blue','brown']


for i in range (5):
    plt.scatter(df1['Assault'][agglo.labels_==i],df['Rape'][agglo.labels_==i],color=colors[i])
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=25,min_samples=5)
dbscan = dbscan.fit(df)
```

```python
dbscan.leaf_size
```

```python
dbscan.labels_
```

```python
colors = ['red','black','orange','blue','brown']


for i in range (5):
    plt.scatter(df1['Assault'][dbscan.labels_==i],df['UrbanPop'][dbscan.labels_==i],color=colors[i])
```

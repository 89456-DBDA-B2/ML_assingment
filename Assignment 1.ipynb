#### Claimants dataset

```python
df4=pd.read_csv("claimants.csv")
print("First 5 rows")
print(df4.head())
print('-' * 150)
print("Last 5 rows")
print(df4.tail())
print('-' * 150)
print("Info display")
print(df4.info())
print('-' * 150)
print("Describe the dataset")
print(df4.describe())
print('-' * 150)
print("Checking null values")
print(df4.isnull().sum())
```

```python
df4.dtypes
```

```python
df4['CLMSEX']=df4['CLMSEX'].fillna(df4['CLMSEX'].mean())
df4['CLMINSUR']=df4['CLMINSUR'].fillna(df4['CLMINSUR'].mean())
df4['SEATBELT']=df4['SEATBELT'].fillna(df4['SEATBELT'].mean())
df4['CLMAGE']=df4['CLMAGE'].fillna(df4['CLMAGE'].mean())
```

```python
print(df4.isnull().sum())
```

- There are no categorical column in Claimants dataset

#### Framingham dataset

```python
df5=pd.read_csv("framingham.csv")
print("First 5 rows")
print(df5.head())
print('-' * 150)
print("Last 5 rows")
print(df5.tail())
print('-' * 150)
print("Info display")
print(df5.info())
print('-' * 150)
print("Describe the dataset")
print(df5.describe())
print('-' * 150)
print("Checking null values")
print(df5.isnull().sum())
```

```python
df5['education']=df5['education'].fillna(df5['education'].mean())
df5['cigsPerDay']=df5['cigsPerDay'].fillna(df5['cigsPerDay'].mean())
df5['BPMeds']=df5['BPMeds'].fillna(df5['BPMeds'].mean())
df5['totChol']=df5['totChol'].fillna(df5['totChol'].mean())
df5['BMI']=df5['BMI'].fillna(df5['BMI'].mean())
df5['heartRate']=df5['heartRate'].fillna(df5['heartRate'].mean())
df5['glucose']=df5['glucose'].fillna(df5['glucose'].mean())
```

```python
print(df5.isnull().sum())
```

- There are no categorical column in framingham dataset

### Q1
- Que1.-
- Predict if a person would buy life insurance based on his age using logistic regression.
- About Dataset - Age: Age of a person
- bought_insurance : whether person buys Insurance or not (0 - No, 1 - Yes)
- predict it for Age : 58,30,20,49,53

- Dataset : insurance_data.csv

- Calculate Accuracy by confusion matrix, Calculate accuracy score, precision score, recall score,f1 score, roc auc score. Also Visualize the result. Also select best parameters using gridsearchCV. 

```python
import warnings
warnings.filterwarnings(action='ignore')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../insurance_data.csv')
```

```python
df.sample(5)
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
sns.distplot(df['age'])
```

```python
X = df[['age']]
y = df[['bought_insurance']]
```

```python
from sklearn.linear_model import LogisticRegression
LoRe = LogisticRegression()
LoRe.fit(X,y)
```

```python
LoRe.score(X,y)
```

```python
y_pred = LoRe.predict(X)
y_pred
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,auc,roc_curve
acc = accuracy_score(y_pred,y)
pre = precision_score(y_pred,y)
rec = recall_score(y_pred,y)
f1 = f1_score(y_pred,y)
# auc = auc(y_pred,y)
# roc_curve=(y_pred,y)
print(acc)
print(pre)
print(rec)
print(f1)
```

```python
plt.plot(df['bought_insurance'],df['age'],label = 'Age')
plt.plot(df['age'],y_pred,color='red')
plt.legend()
```
